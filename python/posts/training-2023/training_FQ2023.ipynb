## **Setting up**

```python
# import some essential data cleaning libraries 
import pandas as pd
import numpy as np
from datetime import datetime

# read in your data 
netflixdata = pd.read_excel("C:\\Users\\betha\\OneDrive\\Desktop\\the-stack\\datasets\\training-2023\\netflixdata2022.xlsx")
# for a csv: pd.read_csv("filename")
```

```python
# see how many rows of data we have
print(netflixdata.shape)
print(len(netflixdata))
```

```python
# print specific columns
netflixdata[['show_id', 'title', 'description']]
```

```python
# check for null values 
print(netflixdata.isnull().values.any())

# check for null values in a specific column
print(netflixdata['director'].isnull().values.any())
print(netflixdata['show_id'].isnull().values.any())
```

```python
# remove all rows with null values
netflixdata_no_null =  netflixdata.dropna()

# looking at new # of rows
print(len(netflixdata_no_null))
```

```python
# check for duplications 
print(netflixdata.duplicated().values.any())
```

```python
# remove all duplicates
netflixdata_no_dups = netflixdata_no_null.drop_duplicates()
```

## **Cleaning, Filtering, and Organizing Data**

```python
# Dataset 1: TV shows released from 2015 - 2020 

# make dataset that is just TV Shows
tv_shows = netflixdata_no_null[netflixdata_no_null['type'] == 'TV Show']

print(len(tv_shows))
```

```python
tv_shows.head(6)
```

```python
# check what type of variable release_year is 
print(tv_shows['release_year'].dtypes)
```

```python
# filter for release years 2015 - 2020
# tv_shows_2015_2020 = tv_shows[(tv_shows['release_year'] <= 2020)  & (tv_shows['release_year'] >= 2015)]
tv_shows_2015_2020 = tv_shows[(tv_shows['release_year'] <= 2020)  & (tv_shows['release_year'] >= 2015)].reset_index()

print(len(tv_shows_2015_2020))

tv_shows_2015_2020.head(6)
```

```python
# drop redundant/inaccurate index column 
tv_shows_2015_2020 = tv_shows_2015_2020.drop(columns = ['index'])

tv_shows_2015_2020.head(6)
```

```python
# getting frequency table for each year 
year_freq = tv_shows_2015_2020.groupby(['release_year']).size()
```

```python
year_freq
```

```python
# Dataset 2: International Movies added to Netflix after 2019

# filter data to just TV shows - another method (less wordy but personally above method is more clear)
movies = netflixdata_no_null.query("type == 'Movie'")
```

```python
movies.head(6)
```

```python
# filter to just International Movies (as categorized by 'listed_in')
movies_international = movies.loc[(movies['listed_in'].str.contains("International Movies") == True)].reset_index() 
movies_international = movies_international.drop(columns = ['index'])

movies_international.head(6)
```

```python
# filtering Intl movies to those added after 2019
# this part is a little trickier!

# check column data type of date_added
print(movies_international['date_added'].dtypes)
```

```python
# convert date_added to datetime and extract the year 
movies_international['date_added'] = pd.to_datetime(movies_international['date_added']).dt.strftime('%Y')
print(movies_international['date_added'].dtypes)
# converting to integer so we can use >= operator
movies_international['date_added'] = movies_international['date_added'].astype(int)
print(movies_international['date_added'].dtypes)
```

```python
# filter to just movies after 2019
movies_international = movies_international[movies_international['date_added'] >= 2019].reset_index()

movies_international.head(6)
```

```python
# Dataset 3: calculate average duration of movies grouped by year

# check column data type of duration
print(movies['duration'].dtypes)

# remove ' min' from duration column
movies['duration'] = movies['duration'].str.replace(' min', '')

# convert duration to integer
movies['duration'] = movies['duration'].astype(int)

movies.head(6)
```

```python
# calculate average duration of movies grouped by year
avg_duration = movies.groupby(['release_year'])['duration'].mean().reset_index()

avg_duration
```

```python
# export dataframe to csv

avg_duration.to_csv('avg_duration.csv', index = False)
```
