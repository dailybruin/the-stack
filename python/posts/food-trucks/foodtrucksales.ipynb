```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
# weeklysales = pd.read_csv("foodtrucks_weektotal.csv")
weeklysales = pd.read_csv("weekly_swipes.csv")

weeklysales.head(3)

# Creating the x labels
weeklysales["X_Label"] = weeklysales["quarter"] + weeklysales["week"].astype(str) +" "+ weeklysales["year"].astype(str)

# Plotting the data
plt.figure(figsize=(15,6))
plt.plot(weeklysales["X_Label"], weeklysales["swipes"], marker='o', linestyle='-')

# Formatting the graph
plt.xlabel("Quarter and Week")
plt.ylabel("Weekly Total")
plt.title("Weekly Total Over Time")
plt.xticks(rotation=90)
plt.grid()

# Show the plot
plt.show()
```

```python
# by quarter
weeklysales_filtered = weeklysales[weeklysales["quarter"] != "B"] # exclude the B aka break

quarters = weeklysales_filtered["quarter"].unique()

# Plotting each quarter separately
for quarter in quarters:
    subset = weeklysales_filtered[weeklysales_filtered["quarter"] == quarter]
    plt.plot(subset["week"], subset["swipes"], marker='o', linestyle='-', label=f"quarter {quarter}")

# Formatting the graph
plt.xlabel("Week")
plt.ylabel("Weekly Total")
plt.title("Weekly Total Over Time by Quarter")
plt.legend(title="Quarter")
plt.grid()

# Show the plot
plt.show()
```

```python
# Load dataset
weeklysales_day = pd.read_csv("foodtruckdailysales.csv")

weeklysales_day.head(3)

weeklysales_day = weeklysales_day.drop(weeklysales_day.tail(1).index) # drop last row

# Creating the x labels
weeklysales_day["X_Label"] = weeklysales_day["Date"].astype(str)

# Plotting the data
plt.figure(figsize=(30,10))
plt.plot(weeklysales_day["X_Label"], weeklysales_day["Daily_Total"], marker='o', linestyle='-')

# Formatting the graph
plt.xlabel("Day")
plt.ylabel("Daily Total")
plt.title("Daily Total Over Time")
plt.xticks(rotation=90)
plt.grid()

# Show the plot
plt.show()
```
