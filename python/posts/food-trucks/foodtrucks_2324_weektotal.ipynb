```python
import pandas as pd

# Load the Excel file
file_path = 'Responsive Documents (25-4753) (1).xlsx'
df = pd.read_excel(file_path)

# Display the first few rows to understand the structure
print("Data preview:")
print(df.head())

# Check column names
print("\nColumn names:")
print(df.columns)

# Assuming relevant columns are 'Quarter', 'Week', and 'Swipes'
# Adjust based on actual column names
# Let's try to infer the correct names for the aggregation step

# Attempt to normalize column names (strip whitespace, lowercase)
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

# Look for candidate columns for quarter, week, and swipes
print("\nNormalized column names:")
print(df.columns)

# Assuming column names like 'quarter', 'week', and 'swipes' now exist
# You may need to adjust these names depending on the actual column titles

grouped = df.groupby(['year', 'quarter', 'week'])['swipes'].sum().reset_index()
grouped = grouped.sort_values(by=['year', 'quarter', 'week'])

grouped = grouped[['swipes', 'quarter', 'week','year']]

# Output the result
print("\nTotal meal swipes per week:")
print(grouped)

# Save to CSV if needed
grouped.to_csv('weekly_swipe_totals.csv', index=False)
print("\nWeekly totals saved to 'weekly_swipe_totals.csv'")
```
