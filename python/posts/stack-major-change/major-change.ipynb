```python
import csv

students = []
with open('data.csv', mode = 'r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        new_dict = {
            'start_major' : row['start_major'],
            'end_major' : row['end_major']
        }
        students.append(new_dict)

students
```

```python
# Initialize a dictionary to store counts for each major transition
major_transitions = {}
# Define all possible majors
# majors = ['major1', 'major2', 'major3', 'major4', 'major5', 'major6', 'major7', 'major8', 'major9', 'major10']
# Initialize counters for each major transition
for start_major in majors:
    for end_major in majors:
        major_transitions[(start_major, end_major)] = 0
# Iterate over students
for student in students:
    start_major = student['start_major']
    end_major = student['end_major']
    major_transitions[(start_major, end_major)] += 1
```
