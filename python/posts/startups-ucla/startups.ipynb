```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/Users/joeyling/Downloads/animated-treemap/files/founders.csv')
df
```

```python
df["Age"] = 2023-df["Year Founded"]
df
```

```python
df["Stage"].unique()
```

```python
def text_to_funding(x):
    if x == 'Late (25K+ MRR, 1M +Funding) ':
        return 3
    elif x== 'Mid (1-25K MRR, 50k-1M Funding)':
        return 2
    else:
        return 1
```

```python
df["Funding"] = df["Stage"].apply(text_to_funding)
```

```python
df
```

```python
plt.bar(df["Age"], df["Funding"])
```

```python
fig, ax = plt.subplots(1)
def plot_hist(df):
    ax.hist(df['Age'], alpha = 0.5)
df.groupby("Funding").apply(plot_hist)
```

```python
industry = df["Industry"].dropna()
industry.value_counts()
```

```python
industries = []
for i in industry:
    ind = i.split(",")
    for j in ind:
        industries.append(j)
print(len(industries))
```

```python
industries = np.array(industries)
#note one startup can belong to multiple industries
np.unique(industries)
#startups within each industry for each year, animated decision tree
#pop within each region for each state
```

```python
df["Age"].value_counts()
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))
plt.hist(industries)
plt.xticks(rotation=40)
plt.show()
```
