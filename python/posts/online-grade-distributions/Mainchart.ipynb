```python
pre = pd.read_csv('19F.csv', encoding="ISO-8859-1" )
post = pd.read_csv('20F.csv',encoding="ISO-8859-1")
```

```python
pre
```

```python
pre['GRADE'].unique()
```

```python
grades = ['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D','D-',
       'F']
```

```python
grades
```

```python
amounts = []
for grade in grades:
    num = pre['GRD_COUNT'][pre['GRADE']==grade].sum()
    amounts.append(num)
```

```python
nums = np.array(amounts)
```

```python
pcts = nums/nums.sum()
```

```python
pcts
```

```python
amountspost = []
for grade in grades:
    num = post['GRD_COUNT'][post['GRADE']==grade].sum()
    amountspost.append(num)
```

```python
numspost = np.array(amountspost)
```

```python
pctspost = numspost/numspost.sum()
```
