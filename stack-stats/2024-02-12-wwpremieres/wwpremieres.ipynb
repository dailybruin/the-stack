```python

import pandas as pd
import numpy as np

wwpremieres = pd.read_csv("westwood_movie_premieres.csv")
wwpremieres
```

```python
import matplotlib.pyplot as plt

movies = [2, 2, 4, 1, 4, 1, 0, 0, 0, 0, 2, 1]
shows = [1, 1, 3, 1, 0, 2, 0, 0, 0, 0, 1, 0]
months = ['Jan.', 'Feb.', 'March', 'April', 'May', 'June', 'July', 'Aug.', 'Sept.', 'Oct.', 'Nov.', 'Dec.']

#fig = plt.figure(1, [8, 7])
fig, ax = plt.subplots(figsize = (8, 7))



movBar = ax.bar(months, movies, width = 0.75, color = 'gold', label = 'Movie')
showBar = ax.bar(months, shows, bottom = movies, width = 0.75, color = 'royalblue', label = 'TV Show')

ax.bar_label(movBar, label_type = 'center', labels=[f'{v:d}' if v != 0 else '' for v in movies])
ax.bar_label(showBar, label_type = 'center', labels=[f'{v:d}' if v != 0 else '' for v in shows])

plt.rcParams['font.size'] = 12
plt.legend(fontsize = 14)
title_font = {'fontname':'Roboto', 'fontsize': 16, 'fontweight': 'bold'}
y_font = {'fontname':'Roboto', 'fontsize': 12}
plt.ylabel('Number of Premieres', **y_font)
plt.title('Movie and TV Show Premieres in Westwood', **title_font)
```
