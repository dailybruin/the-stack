```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
plt.rcParams["font.family"] = "monospace"
```

```python
boba = pd.read_csv('stackstats.boba data  - reorganized.csv')
```

```python
boba.head
```

```python
boba_type = boba[['Store', 'Milk Tea', 'Tea/Fruit Tea', 'Blended', 'Specialty']]
```

```python
boba_type.plot(x='Store', kind='bar', stacked=True, 
               figsize=(8,5), color=['#A27A6B', '#D69581', '#714D3E', '#391403'], 
               title='Menu Variety of Westwood Boba Shops')
plt.xlabel("Shop");
plt.ylabel("Number of Drinks");
handles, labels = plt.gca().get_legend_handles_labels()
order = [3,2,1,0]
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order])
plt.xticks(rotation=25);
plt.show()
```
