```python
import pandas as pd
import numpy as np
```

```python
data = []
institutions = ["UCLA", "Stanford", "USC"]
years = range(2013, 2024)

for institution in institutions:
    for year in years:
        data.append({"Institution": institution, "Year": year, "Championships": 0})

df = pd.DataFrame(data)

# Replace the "Championships" column with your custom data
ucla_championships = [110, 112, 112, 113, 113, 116, 116, 118, 119, 120, 121]  # Make sure the length is 11
stanford_championships = [99, 105, 107, 109, 112, 119, 119, 126, 128, 131, 134]  # Make sure the length is 11
usc_championships = [104, 100, 100, 102, 103, 107, 107, 107, 110, 111, 112]  # Make sure the length is 11

df.loc[df["Institution"] == "UCLA", "Championships"] = ucla_championships
df.loc[df["Institution"] == "Stanford", "Championships"] = stanford_championships
df.loc[df["Institution"] == "USC", "Championships"] = usc_championships

df
```

```python
# Make a lineplot using seaborn library, with Year as x-axis, and Championships as y-axis

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")
sns.lineplot(data=df, x="Year", y="Championships", hue="Institution", palette=['#2D68C4', '#658f7c', '#990000'])

# Rename y-axis label to "Total Team Championships"
plt.ylabel("Total Team Championships")

# Display all years
plt.xticks(np.arange(min(df["Year"]), max(df["Year"])+1, 1.0))
# Tilt x-axis labels
plt.xticks(rotation=45)

# plt.show()

# Export as png
plt.savefig("ncaachamps.png", transparent=True, dpi=300)
```
