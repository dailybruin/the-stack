```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ai = pd.read_csv('ai.csv')
```

```python
ai
```

```python
plt.rcParams.update({'font.size': 17})

plt.figure(figsize=(15, 12))
barplot = sns.barplot(x='days', y='online_service', data=ai, orient='h', color='skyblue')

for i in range(len(ai)):
    if i == len(ai) - 1:  # if this is the last bar
        barplot.text(ai['days'][i] - 8, i, ai['label'][i], va='center', ha='right')
    elif i == 1:  # if this is the second bar
        barplot.text(ai['days'][i] + 8, i, ai['label'][i], va='center', fontweight='bold')
    else:
        barplot.text(ai['days'][i] + 8, i, ai['label'][i], va='center')

# Increase title size and make it bold
plt.title('Time taken to reach 1 million uses (days)', fontsize=25, fontweight='bold')
plt.xlabel('Days', fontsize=20, fontweight='bold')
plt.ylabel('Online Service', fontsize=20, fontweight='bold')

# Make the label of the second item in the y-axis bold
labels = barplot.get_yticklabels()
if len(labels) > 1:  # if there is more than one label
    labels[1].set_fontweight('bold')

# Remove tick marks
plt.tick_params(bottom=False, left=False)

plt.show()
```

```python


# Sample DataFrame
data = {'Category': ['A', 'B', 'C', 'D', 'E'],
        'Value': [10, 15, 7, 10, 5],
        'Labels': ['label1', 'label2', 'label3', 'label4', 'label5']}
df = pd.DataFrame(data)

# Create bar plot
plt.figure(figsize=(10,6))
barplot = sns.barplot(x='Category', y='Value', data=df)

# Add labels to each bar
for i in range(df.shape[0]):
    barplot.text(i, df.Value[i] + 0.5, df.Labels[i], ha = 'center')

plt.show()
```
