```python
import pandas as pd
import json
dataUCB = pd.read_csv ("./ucb-prof-donations.csv")
newlistUCB = dataUCB['committee_name'].drop_duplicates().to_list()

```

```python
final_dataUCB = []
for x in newlistUCB: 
    final_dataUCB.append(dataUCB[dataUCB.committee_name == x].contribution_receipt_amount.sum())
```

```python
df  = pd.DataFrame()
df['committee_name'] = newlistUCB
df['contribution_receipt_amount']  = final_dataUCB
topFive = df.sort_values('contribution_receipt_amount', ascending = False).head()
```

```python
y = json.dumps(topFive)

print(y)
```
